```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(2+x)
```

```python
xs = np.linspace(0, 4, 400)
ys =f(xs)
plt.plot(xs, ys, label=r"$y=\sqrt{x+2}$")
plt.plot(xs, xs, label=r"$y=x$")
plt.legend()
```

```python
def suite(u0, n):
    resultat = list()
    resultat.append(u0)
    for _ in range(n):
        resultat.append(f(resultat[-1]))
    return resultat
    
```

```python
suite(0.5, 10)
```

```python
suite(3.5, 10)
```

```python
def visualisation( n):
    us = suite(0.5, n)
    u2 = suite(3.5,n)
    double = [u for u in us for _ in range(2)]
    plt.plot(double[:-1], double[1:], marker = ".")
    double2 = [u for u in u2 for _ in range(2)]
    plt.plot(double2[:-1], double2[1:], marker = ".")
    xs = np.linspace(0, 4, 400)
    ys =f(xs)
    plt.plot(xs, ys, label=r"$y=\sqrt{x+2}$")
    plt.plot(xs, xs, label=r"$y=x$")
    plt.legend()
```

```python
visualisation( 10)
```

```python
def suite2(u0, n):
    resultat = list()
    resultat.append(u0)
    for _ in range(n):
        resultat.append(np.exp(-resultat[-1]))
    return resultat
```

```python
suite2(0.5,10)
```

```python
def visualisation2(u0, n):
    us = suite2(u0, n)
    double = [u for u in us for _ in range(2)]
    plt.plot(double[:-1], double[1:], marker = "o")
    xs = np.linspace(0.49, 0.62, 400)
    ys =np.exp(-xs)
    plt.plot(xs, ys, label=r"$y=e^{-x}$")
    plt.plot(xs, xs, label=r"$y=x$")
    plt.legend()
```

```python
visualisation2(0.5,10)
```

```python
def yn(u0, n):
    l = suite(u0,n+2)
    resultat = list()
    for i in range(n):
        resultat.append((l[i+2]-l[i+1])/(l[i+1]-l[i]))
    return resultat
```

```python
yn(0.5, 28)
```

```python
def yn2(u0, n):
    l = suite2(u0,n+2)
    resultat = list()
    for i in range(n):
        resultat.append((l[i+2]-l[i+1])/(l[i+1]-l[i]))
    return resultat
```

```python
yn2(0.5,25)
```

```python
def loga(u0, n):
    return [np.log(np.abs(x-2)) for x in suite(u0,n)]
```

```python
loga(0.5,20)
```
